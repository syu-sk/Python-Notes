Displaying full data:
- It is important to note that setting the max number of columns/rows any number below the actual number of columns/rows in the dataframe will still result in the dataframe displaying its previewed version.

```python
pd.set_option('display.max_columns', 114)
#.set_option('display.max_columns', x), where x is number of columns. response_data has 114 columns
pd.set_option('display.max_rows', 87)
#.set_option('display.max_rows', y), where y is number of rows. data has 87 columns
```
