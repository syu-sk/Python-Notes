Grouping:
- Works by isolating a group of data to analyse it (by country, age group, school et cetera)

```python
response_df['MainBranch'].value_counts()
```

```python
response_df['EdLevel'].value_counts()
```

Let's say the relationship between education level & reason for coding and age is to be investigated.

```python
grp1 = response_df.groupby(['MainBranch', 'EdLevel'])
#returns 'groupby' object
```

```python
grp1.get_group(('I am learning to code', 'Bachelor’s degree (B.A., B.S., B.Eng., etc.)'))
#works similar to a filter, but it is way easier to filter by multiple criteria
```

Apart from its .get_group method acting as a multi-filter, the groupby method does not isolate one type, but rather groups each type with each other. When the groupby object is then passed into a function, the function is executed on each of these 'groups'.

```python
grp1['Age'].value_counts().head(100)
```

- This code above has 2 groups, 'MainBranch' and 'EdLevel'. Calling the column value of 'WorkExp' will return everybody under each education level, under each mainbranch. The first value inside the list from .groupby appears to be the one with the larger bracket. i.e largest hierarchial index
- Rather than a filter which would only give responses from a single edlevel and mainbranch, running through groupby will give responses from every edlevel and mainbranch, grouped accordingly.
- Essentially does something that would take filters for every single ed level and mainbranch.

To simplify further examples, let's use a single group.

```python
educationgrp = response_df.groupby('EdLevel')
```

```python
educationgrp.get_group('Bachelor’s degree (B.A., B.S., B.Eng., etc.)').head(3)
#similarly, all entries will have a bachelor's degree education level
```

Getting aggregate statistics:

```python
educationgrp['WorkExp'].median()
#in statistics, often median is used in place of mean to reduce the effect of outliers
#series is returned (1 index: 1 value)
```

Grouped by education level. Thus when median work experience is called, it returns the median work experience FOR EACH education level

```python
educationgrp['WorkExp'].agg(['median', 'mean'])
#.agg required for multiple aggregate values
#dataframe created since it is not one-dimensional
```

Finding values:

```python
educationgrp.get_group('Bachelor’s degree (B.A., B.S., B.Eng., etc.)')['RemoteWork'].str.contains('Hybrid')
#.value_counts to get number of Hybrids 
```

The above gets all indexes which have a bachelor's degree education level. 

The 'RemoteWork' column is then called for all indexes (which thus returns a series).
