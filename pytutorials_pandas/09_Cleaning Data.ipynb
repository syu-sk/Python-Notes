```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
#uses a different sample dataframe with intended lost/missing values
```

Missing values include: Strings[NA, Missing], NaN values from numpy, None values from Python etc.

```python
df = pd.DataFrame(people)
df
```

***
Handling Missing Values:

1. Missing values are dropped

```python
df.dropna()
#drops INDEXES which have NaN or None values in them
```

Default arguments are axis='index', and how = 'any'.

Axis defines whether the index or column of the null value is dropped.

How defines the conditions for the drop. 'all' requires all values in the dropping axis to be null values for the axis to be dropped

```python
df.dropna(axis='columns', how='all')
#no columns dropped as all of them have at least one value that is not null
```

If the dropping axis is only to be dropped if a value in certain non-dropping axes is null only, a subset parameter can be passed. (For example, drop row only if email is NaN)

The how parameter works similarly, with 'all' meaning every subset has to be a null value to be dropped.

```python
df.dropna(axis='index', how='any', subset=['email'])
#how parameter does not have significance if there is only 1 subset
#subset can have multiple columns
```

2. Replacing missing values with NaN values

```python
df.replace({'NA': np.nan, 'Missing': np.nan}, inplace=True)
```

```python
df
```

Now, these NaN values can be directly dropped.

```python
df.dropna()
```

```python
df.isna()
#handy method for identifying NaN values -- NaN = True
```

3. Filling NaN values

```python
df.fillna('NONINPUT')
#works with strings and integers
```

***
Casting Datatypes:
- When using aggregation, values have to be int/float. However data may be imported in the form of strings.
- Hence data should be casted into int/float datatypes

```python
df.replace({'NA': np.nan, 'Missing': np.nan}, inplace=True)
df.fillna(np.nan)
```

```python
#df['age'] = df['age'].astype(int)
#.astype also works on dataframe objects
```

The above is the usual way to cast datatypes. However this will return an exception if any of the values in the series is uncastable into an integer. (like NaN values)

There are generally 2 options:
- Convert i.e replace the uncastable values into something else that is castable (0 etc.)
- Choose another type to cast to

```python
df['age'] = df['age'].astype(float)
#uses the second option to cast to float, which accepts NaN values
```

```python
df['age'].median()
```
