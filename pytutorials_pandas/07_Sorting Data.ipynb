**07: Sorting Data**
- Sort Columns (rearrange rows)
- Grab smallest/largest values from each column

```python
import pandas as pd
```

```python
students = {
    'names': ['Tom', 'Bob', 'Jane', 'May'],
    'age': [9, 10, 10, 9],
    'subjects': ['Science', 'Arts', 'Hybrid', 'Arts'],
    'award winner': [True, False, False, True]
}
```

```python
df = pd.DataFrame(students)
df
```

***
Sorting Columns:
1. .sort_values
   - 'by' parameter, with a specified column has to be passed.
   - Accepts ascending parameter. If set to False, reverses the indexes
   - Accepts inplace parameter
2. .sort_index
   - Sorts by index, default ascending.
   - Accepts ascending parameter

```python
df.sort_values(by='names')
```

If there are duplicate values, the duplicates may be then sorted amongst each other, by passing a list into the by parameter. These values can then be configured to be sorted ascending/descending, by also passing a list into the ascending parameter.

```python
df.sort_values(by=['age', 'names'], ascending=[True, True], inplace=True)
#names -- May -> Tom, Bob -> Jane
df
```

```python
df.sort_index()
```

***
Sorting a single column:
1. .sort_values() - Method used for a pd Series

```python
df['award winner'].sort_values()
#returns a series. 
```
