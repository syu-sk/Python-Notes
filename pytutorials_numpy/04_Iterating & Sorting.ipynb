# 04: Iterating and Sorting
<br>

`numpy` has its unique way of iteration and sorting, providing easier options to manipulate arrays.

```python
import numpy as np
```

***
## Iteration

#### 1-D 

```python
np1 = np.array([1,2,3,4,5,6])
for i in np1: 
    print(i) #simple, as with lists
```

#### 2-D

```python
np2 = np1.reshape(2,3)
for i in np2:
    print(i)
```

This simply returns the (n-1)th dimension in the form of a series.
<br>
But what if the intended output was the individual objects? Certainly, we could nest loops.

```python
for i in np2:
    for num in i:
        print(num)
```

But for large n-dimensional arrays, this may not be so easy. 
- For individual integers, this can be done easily by flattening the array.

```python
for i in np2.reshape(-1):
    print(i)
```

#### nditer

Even simpler, `numpy` has its own dedicated iterator, `np.nditer`.

```python
for i in np.nditer(np2):
    print(i)
```

***
## Sorting
<br>

`numpy` has its own `sort` function as well.

#### 1-D

```python
np3 = np.array([6,3,7,1,5,4,8])
print(np.sort(np3))
print(np3)

np4 = np.array(['John', 'Bob', 'Alice'])
print(np.sort(np4))
print(np4)

np5 = np.array([True, False, True, False])
print(np.sort(np5))
print(np5)
```

It is important to note that `.sort` only creates a temporary copy (unless assigned) at the moment it is called. It does not change the original array.

#### 2-D

```python
np6 = np.array([[7,3,9,0], [1,4,8,2]])
print(np.sort(np6))
```

For multidimensional arrays, `.sort` sorts within the group.
