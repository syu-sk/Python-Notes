# 02: Universal Functions
<br>

This chapter contains a bulk of the functions used to manipulate `ndarrays`

```python
import numpy as np
```

***
#### Function 1: sqrt

```python
np1 = np.reshape([1,4,9,16], [2,2])
print(np1)
print(np.sqrt(np1)) #evidently, it works on n-dimensional arrays as well
```

***
#### Function 2: absolute

```python
np2 = np.reshape([-2,-1,0,1], [2,2])
print(np2)
print(np.absolute(np2))
```

***
#### Function 3: exp (for Exponential)
Substitutes each integer into e^x, then returns the result as output

```python
print(np.exp(np2))
```

***
#### Function 4: min/max

```python
print(np.max(np2), np.min(np2))
```

***
#### Function 5: sign
Returns -1 for each negative value, 1 for positive value

```python
print(np.sign(np2))
```

***
#### Function 6: sin, cos, tan
Applies trigonometry functions to each element

```python
np3 = np.array([1, 0, -1])
print(np.sin(np3))
print(np.cos(np3))
print(np.tan(np3))
```

There are all of the typical mathematical functions for `ndarrays`.
