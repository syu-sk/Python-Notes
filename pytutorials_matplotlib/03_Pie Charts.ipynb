**03: Pie Charts**

```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
```

***
Making Pie Charts:
- Use plt.pie method. Note that the proportions do not necessarily have to add up to 100, proportions are calculated automatically.

```python
slices = [20, 30, 40, 10]
labels = ['slice 1', 'slice 2', 'slice 3', 'slice 4']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})
#plt.pie(proportions, labels). wedgeprops parameter to set black border
#check mpl documentation for more

plt.tight_layout()
#get rid of the mpl object text
```

***
Real life section:

Below is a chunk of code which creates two lists, one for the 15 most common programming languages known, and the other for the number of people who know each language

Data is based on a Stack Overflow dev survey, mpldata.csv under the same directory

```python
df = pd.read_csv('02_03_mpldata.csv')
ids = df['Responder_id']
langs = df['LanguagesWorkedWith']

lang_counter = Counter()
for response in langs:
    lang_counter.update(response.split(';'))

lang_x = [i[0] for i in lang_counter.most_common(15)]
number_y = [i[1] / 1000 for i in lang_counter.most_common(15)]
```

```python
plt.title('15 Most Common Programming Languages')

plt.pie(number_y, labels=lang_x, wedgeprops={'edgecolor':'black'})
plt.tight_layout()
```

Of course, pie charts are not too descriptive of data, mostly used for presentations etc.

Especially in this case where there is so much data, a pie chart may often be confusing rather than make it clearer

***
For more functionality with matplotlib, check documentation. Other useful functionality include:
1. explode parameter: separates a specified chunk of the pie chart to highlight a portion of it
2. shadow parameter: gives the pie a 'shadow' effect; makes it look more 3d
3. startangle parameter: defines where the pie chart starts forming. default is 0, starting from east. goes anti-clockwise
4. autopct='%1.1f%%' parameter: displays the percentage of the chart that each pie takes up
