**07: Plotting with Datetime Objects**

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
y = [0, 1, 3, 4, 6, 5, 7]
```

***
Plotting datetime:
- plt.plot_date(dates, y)
- Accepts general plot parameters (color, label, linewidth etc.)

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()
#declutters the x axis - gcf for get current figure. autofmt for autoformat|

plt.tight_layout()
```

Date Format:
- Similar to changing the date format in the datetime module. Numerical, alphabetical, full form etc.
- All formats can be found under the datetime module documentation

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
#reformats x axis. gca for get current axis (where x is then specified). set format as above

plt.tight_layout()
```

***
Real life section:

```python
data = pd.read_csv('07_mpldata.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
#good practice to sort dates - this uses pandas, nothing related to mpl

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.tight_layout()
```
