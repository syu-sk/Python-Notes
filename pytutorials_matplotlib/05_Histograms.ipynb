**05: Histograms**
- Effective for visualising distribution of data, when within a boundary.
- Groups data into bins, as opposed to bar graphs which plot according to an individual value

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
```

***
Creating Bins:
- These are essentially the brackets that x values will be grouped into.
- Bins are created via:
  1. Passing an integer - integer will be the number of bins created
  2. Passing a list
- General plot parameters accepted, but edgecolor is advised to keep the plot unambiguous

```python
plt.hist(ages, bins=5, edgecolor='black')
plt.title('Ages')
```

Passing a list will specify the x values at which the edge of the bins are.
- Typically, this is the preferred method since the bins are at regular values

```python
plt.hist(ages, bins=[10, 20, 30, 40, 50, 60], edgecolor='black')
```

***
Real life section:

```python
data = pd.read_csv('05_mpldata.csv')
ids = data['Responder_id']
ages = data['Age']
```

```python
bins = [10*i for i in range(11)]
plt.hist(ages, bins=bins, edgecolor='black')
```

Plotting on a log scale is also possible by passing True in the log parameter.

Provides easier comparisons between bins with extreme values

```python
plt.hist(ages, bins=bins, edgecolor='black', log=True)
```
