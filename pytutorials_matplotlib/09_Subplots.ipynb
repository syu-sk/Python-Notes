**09: Subplots**

Recall that in all prior videos, all plots have been made via plt.plot... or the like. This also means that any and all axes (plots) made on the figure (graph) must be done in the same cell.

However, this is not necessarily true for mpl. At its root it is still object-oriented.

```python
import pandas as pd
from matplotlib import pyplot as plt
```

***
Figures and axes introduction:
- Creates a pyplot instance

```python
data = pd.read_csv('09_mpldata.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

1 axe on 1 figure:

```python
fig, ax = plt.subplots()

ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')

ax.legend()
ax.set_title('Median Salary (USD) by Age')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()
#think of tight_layout as a class method
```

2 axes on 1 figure:
- Similar to the next example - however there is only 1 figure i.e both plots are in the same png
- Each axe can have a different title, x label, y label etc. tight_layout remains as a plt class method and only has to be passed once
- sharey=True enables the y-scale for both plots in the same figure to be corrected to each other to compare easier. Works the same for x

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True)

ax1.plot(ages, py_salaries, label='Python')
ax1.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
```

2 axes on 2 figures:
- 2 axes (plots), 1 with 2 graphs and the other with 1, split across 2 figures

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, py_salaries, label='Python')
ax1.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
```
