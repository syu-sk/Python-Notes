**06: Scatter Plots**
- Shows relationship between two sets of values to find correlation
- Recall linearly related, non-linearly related, r value etc.

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

***
Making a scatter plot:
- plt.scatter(x axis, y axis). Accepts general plot parameters (color, alpha etc.)

```python
plt.scatter(x, y, s=25, edgecolor='black', linewidth=1, alpha=0.75)
#s for size
plt.tight_layout()
```

Colour Maps:
- Making each plot point have a differing colour, which can then be used to signify something
- To give a plot point a colour, it is as simple as passing a list into the color parameter, where each plot point will have its corresponding colour. Sizes can be changed in this way as well, though size mapping is not a thing
- However, normally for a presentation or while analysing data, the colour often is to be of certain significance, whereby colour maps are used to represent a degree of a characteristic (essentially creating a third axis for the plot)
- As usual, for matplotlib parameters like cmap, edgecolors etc, head to the documentation to see a list of all provided options

```python
a = [1,2,3,4,5]
b = [1,2,3,4,5]
colours = [5,4,5,1,2]

plt.scatter(a, b, c=colours, cmap='Blues', edgecolor='black')
plt.colorbar().set_label('3rd axis')
#edgecolor advised to be specified - the lowest value on the colormap is white

plt.tight_layout()
```

Real life section:
- Youtube top trending videos about 5 years ago

```python
df = pd.read_csv('06_mpldata.csv')
df
```

```python
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='Reds', edgecolor='black', linewidth=1)
plt.colorbar().set_label('Ratio of Likes to Dislikes')
```

A situation like this is where the effect of the log scale would be really exemplified. A single outlier can skew a linear scale and reduce the readability of data significantly.

```python
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='Reds', edgecolor='black', linewidth=1)
plt.colorbar().set_label('Ratio of Likes to Dislikes')

plt.xscale('log')
plt.yscale('log')
```

Conclusions can be made from this visual (high ratio for highly viewed videos etc. or can make for a good infographic during a presentation.

Matplotlib has many available configurations that can be looked up in its documentation. Many styles, color options etc.
